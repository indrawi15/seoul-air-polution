```python
from scipy.stats import zscore
z_scores = df[[ 'SO2', 'NO2', 'O3', 'CO', 'PM10', 'PM2.5']].apply(zscore)
outliers = (z_scores.abs() > 3).any(axis=1)
percentage_outliers = outliers.mean() * 100
percentage_outliers
```

```python
df_cleaned = df[~outliers]

import matplotlib.pyplot as plt
# Define a function to replace outliers with the median
def replace_outliers_with_median(df_cleaned, column_name):
  median = df_cleaned[column_name].median()
  IQR = df_cleaned[column_name].quantile(0.75) - df_cleaned[column_name].quantile(0.25)
  lower_bound = df_cleaned[column_name].quantile(0.25) - 1.5 * IQR
  upper_bound = df_cleaned[column_name].quantile(0.75) + 1.5 * IQR
  df_cleaned[column_name] = df_cleaned[column_name].apply(lambda x: median if x < lower_bound or x > upper_bound else x)

# Replace outliers for each pollutant
for pollutant in ['SO2', 'NO2', 'CO', 'O3', 'PM10', 'PM2.5']:
  replace_outliers_with_median(df_cleaned, pollutant)
```

```python
# Define the function to calculate forecast accuracy metrics
def forecast_accuracy(forecast, actual):
    nrmse = np.sqrt(np.mean(np.square(forecast - actual))) / \
        (actual.max() - actual.min())
    corr = np.corrcoef(forecast, actual)[0,1]
    r2 = r2_score(actual, forecast)
    sd_actual = np.std(actual)
    sd_forecast = np.std(forecast)
    d1 = 1 - np.sum(np.abs(actual - forecast)) / \
        np.sum(np.abs(actual - np.average(actual)) + \
              np.abs(forecast - np.average(actual)))

    result = {'NRMSE': nrmse,
              'd': d1, 'corr': corr, 'Rsq': r2,
              'SD ratio': sd_forecast / sd_actual}
    for key, value in result.items():
        result[key] = np.round(value, 3)
    return result
```
